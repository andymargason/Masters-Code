```python
import pandas as pd
import numpy as np
```

```python
elwha = pd.read_csv('C:/Users/andrewm4/Documents/MastersDatasets/Kirschbaum_Frontal_Periods/elwha_basin_fronts.csv')
hoh = pd.read_csv('C:/Users/andrewm4/Documents/MastersDatasets/Kirschbaum_Frontal_Periods/hoh_basin_fronts.csv')
queets = pd.read_csv('C:/Users/andrewm4/Documents/MastersDatasets/Kirschbaum_Frontal_Periods/queets_basins_front.csv')
quinalt = pd.read_csv('C:/Users/andrewm4/Documents/MastersDatasets/Kirschbaum_Frontal_Periods/quinalt_basin_fronts.csv')
```

```python
drop_clmns = ('00','sum','1','XLONG', 'XLAT','Bains','basins','fid')
```

```python
elwha_drop = elwha.loc[:,~elwha.columns.str.endswith(drop_clmns)]
hoh_drop = hoh.loc[:,~hoh.columns.str.endswith(drop_clmns)]
queets_drop = queets.loc[:,~queets.columns.str.endswith(drop_clmns)]
quinalt_drop = quinalt.loc[:,~quinalt.columns.str.endswith(drop_clmns)]
```

```python
elwha_mean = elwha_drop.mean()
hoh_mean = hoh_drop.mean()
queets_mean = queets_drop.mean()
quinalt_mean = quinalt_drop.mean()
```

```python
df_concat = pd.concat([elwha_mean, hoh_mean, queets_mean, quinalt_mean], axis=1)
```
