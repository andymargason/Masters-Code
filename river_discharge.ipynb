```python
elwha = pd.read_csv('Elwha_mean.discharge_1975-2020.csv', comment= '#', skiprows=2)
hoh = pd.read_csv('Hoh_mean.discharge_1975-2020.csv', comment= '#', skiprows=2)
queets = pd.read_csv('Queets_mean.discharge_1975-2020.csv', comment= '#', skiprows=2)
quinault = pd.read_csv('Quinault_mean.discharge_1975-2020.csv', comment= '#', skiprows=2)
```

```python
elwha['datetime'] = elwha['datetime'].astype('datetime64')
hoh['datetime'] = hoh['datetime'].astype('datetime64')
queets['datetime'] = queets['datetime'].astype('datetime64')
quinault['datetime'] = quinault['datetime'].astype('datetime64')
```

```python
elwha['water_year'] = elwha.datetime.dt.year.where(elwha.datetime.dt.month < 10, elwha.datetime.dt.year + 1)
```

```python
annual_mean_discharge_rate = elwha.groupby('water_year')[['discharge']].mean()
annual_sum_discharge_rate = elwha.groupby('water_year')[['discharge']].sum()
```
