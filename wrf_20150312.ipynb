```python
import netCDF4
from netCDF4 import Dataset
from wrf import getvar, to_np
import numpy as np
import pandas as pd
import xarray as xr
import matplotlib.pyplot as plt
```

```python
# uploading the ncfiles

nc06 = Dataset('C:/Users/Andrew/Downloads/2015120312/wrfout_d4.2015120312.f06.0000')
nc12 = Dataset('C:/Users/Andrew/Downloads/2015120312/wrfout_d4.2015120312.f12.0000')
nc18 = Dataset('C:/Users/Andrew/Downloads/2015120312/wrfout_d4.2015120312.f18.0000')
nc24 = Dataset('C:/Users/Andrew/Downloads/2015120312/wrfout_d4.2015120312.f24.0000')
```

```python
#12 = 12-04 0000
rain06 = getvar(nc06, 'RAINNC')
rain12 = getvar(nc12, 'RAINNC')
rain18 = getvar(nc18, 'RAINNC')
rain24 = getvar(nc24, 'RAINNC')
```

```python
# Get Snow Variables
snow06 = getvar(nc06, 'SNOWNC')
snow12 = getvar(nc12, 'SNOWNC')
snow18 = getvar(nc18, 'SNOWNC')
snow24 = getvar(nc24, 'SNOWNC')
```

```python
# Get Graupel/Hail Variables
graup06 = getvar(nc06, 'GRAUPELNC')
graup12 = getvar(nc12, 'GRAUPELNC')
graup18 = getvar(nc18, 'GRAUPELNC')
graup24 = getvar(nc24, 'GRAUPELNC')
```

```python
# Get Near Surface Air Temperature
atemp06 = getvar(nc06, "T2")
atemp12 = getvar(nc12, "T2")
atemp18 = getvar(nc18, "T2")
atemp24 = getvar(nc24, "T2")

# In celsius
atemp06C = atemp06 - 273.15
atemp12C = atemp12 - 273.15
atemp18C = atemp18 - 273.15
atemp24C = atemp24 - 273.15
```

```python
# Get Surface Skin Temperature
stemp06 = getvar(nc06, "TSK")
stemp12 = getvar(nc12, "TSK")
stemp18 = getvar(nc18, "TSK")
stemp24 = getvar(nc24, "TSK")

# In Celsius
stemp06C = stemp06 - 273.15
stemp12C = stemp12 - 273.15
stemp18C = stemp18 - 273.15
stemp24C = stemp24 - 273.15
```

```python
da = stemp24C.isel(south_north = slice(230,270), west_east = slice(110,180))
da.plot(cmap = plt.cm.Oranges, levels = [0, 2, 4,6,8,10,12,14,16])
plt.title('Surface Skin Temp Olympics 2015-12-04 0500')
plt.savefig('2015120312/Pictures/Stemp24C_olympics.jpeg')
plt.show()
```

```python
da_elw = stemp24C.isel(south_north = slice(235,277), west_east = slice(160,191))
da_elw.plot(y = 'XLAT', x = 'XLONG', cmap = plt.cm.Oranges, levels = [0, 2, 4,6,8, 10])
plt.xticks(rotation=45)
plt.title('Surface Skin Temp Elwha 2015-12-04 0500')
#plt.savefig('2015120312/Pictures/Stemp24C_elwha.jpeg')
plt.show()
```

```python
graup_tots = graup24.isel(south_north = slice(190,300), west_east = slice(110,230))

graup_tots.plot(y = 'XLAT', x = 'XLONG', cmap = plt.cm.Greens, robust = True)
plt.xticks(rotation=45)
plt.title('Graupel over the Olympics 2015-12-04 0500')
#plt.savefig('2015120312/Pictures/Graup24_olympics.jpeg')
plt.show()
```

```python
graup_elwa = graup24.isel(south_north = slice(243,253), west_east = slice(180,190))

graup_elwa.plot(y = 'XLAT', x = 'XLONG', cmap = plt.cm.Greens)
plt.xticks(rotation=45)
plt.title('Graupel over Eastern Side of the Elwha 2015-12-04 0500')
plt.savefig('2015120312/Pictures/Graup24_elwha.jpeg')
plt.show()
```

```python
plt.hist(graup_elwa, color = ('green','green','green','green','green','green','green','green','green','green',))
plt.title('Graupel Distribution Over the Elwha 2015-12-04 0500')
plt.xlabel("Precipitation (mm)")
plt.ylabel("Frequency")
plt.savefig('2015120312/Pictures/Graup24_elwha_hist.jpeg')
plt.show()
```
